Simulaciones de base para comparación con métodos de redes neuronales en la predicción de precio de criptomonedas.

```python
# Manipulación de datos
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt # Plots
params = {'legend.fontsize': 'x-large',
         'axes.labelsize': 'x-large',
         'axes.titlesize':'x-large',
         'xtick.labelsize':'x-large',
         'ytick.labelsize':'x-large'}
plt.rcParams.update(params)

import statsmodels.api as sm
from statsmodels.tsa.api import ExponentialSmoothing, SimpleExpSmoothing, Holt

# Keras
from math import sqrt
from sklearn.metrics import mean_squared_error
```

```python
df = pd.read_csv('data/data.csv', usecols=['Date','bitcoin'])
df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date', inplace=True)

plt.figure(figsize=(20,5))
plt.title("Precio Bitcoin", fontsize=25)
plt.plot(df, label="bitcoin")
plt.legend()
plt.show()

test_size=30
window_size=1
dia_futuro=[1,2,5,7]

train = df[0:len(df)-test_size]
test  = df[len(df)-test_size:]
plt.figure(figsize=(20,5))
plt.title("Training/Testing", fontsize=25)
plt.plot(train, label="train")
plt.plot(test, label="test")
plt.legend()
plt.show()
```

Método "naive": el precio del día siguiente es igual al anterior -> y(t+1) = y(t)

```python
#Método
y_naive = test.copy()
for f in dia_futuro:    
    naive = np.asarray(df[len(train)-f:len(train)-f+test_size].bitcoin)
    y_naive['naive_' + str(f)] = naive

#Graficar la serie completa
plt.figure(figsize=(20,8))
plt.plot(train.index, train['bitcoin'], label='Train')
plt.plot(test.index, test['bitcoin'], label='Test')
for i in range(1,len(y_naive.columns)):
    plt.plot(y_naive.index, y_naive[y_naive.columns[i]], label=y_naive.columns[i])
plt.legend(loc='best')
plt.title("Predicción precio Bitcoin (Naive)",fontsize=20)
plt.show()

#Graficar solo la porción de test
plt.figure(figsize=(20,8))
plt.plot(test.index, test['bitcoin'], label='Test')
for i in range(1,len(y_naive.columns)):
    plt.plot(y_naive.index, y_naive[y_naive.columns[i]], label=y_naive.columns[i])
plt.legend(loc='best')
plt.title("Predicción precio Bitcoin (Naive) zoom",fontsize=20)
plt.show()

#Cálculo de error
rmses_y = []
rmses_x = []
for i in range(1,len(y_naive.columns)):
    rmse = sqrt(mean_squared_error(test.bitcoin, y_naive[y_naive.columns[i]]))
    print("RMSE " + y_naive.columns[i] + ": " + str(rmse))
    rmses_y.append(rmse)
    rmses_x.append(y_naive.columns[i])  
    
#Graficar RMSE
width = 1/1.5
plt.figure(figsize=(20,5))
plt.bar(rmses_x, rmses_y, width)
plt.xticks(rotation=20)
plt.title("RMSE",fontsize=20)
plt.show()  
```

Se puede ver por el RMSE y el gráfico, que el método no es adecuado para conjuntos de datos con alta variabilidad.

Método "Promedio simple": el precio del día siguiente es igual al promedio de los días previamente observados.

```python
#Método
y_promedio = test.copy()
for f in dia_futuro:
    y_promedio['promedio_' + str(f)] = np.nan
    for i in range(0,test_size):    
        training = np.asarray(df[0:len(df)-test_size+i-f+1])
        y_promedio['promedio_' + str(f)][i] = training.mean()

#Graficar la serie completa
plt.figure(figsize=(20,8))
plt.plot(train.index, train['bitcoin'], label='Train')
plt.plot(test.index, test['bitcoin'], label='Test')
for i in range(1,len(y_promedio.columns)):
    plt.plot(y_promedio.index, y_promedio[y_promedio.columns[i]], label=y_promedio.columns[i])
plt.legend(loc='best')
plt.title("Predicción precio Bitcoin (Promedio)",fontsize=20)
plt.show()

#Graficar solo la porción de test
plt.figure(figsize=(20,8))
plt.plot(test.index, test['bitcoin'], label='Test')
for i in range(1,len(y_promedio.columns)):
    plt.plot(y_promedio.index, y_promedio[y_promedio.columns[i]], label=y_promedio.columns[i])
plt.legend(loc='best')
plt.title("Predicción precio Bitcoin (Promedio) zoom",fontsize=20)
plt.show()

#Cálculo de error
rmses_y = []
rmses_x = []
for i in range(1,len(y_promedio.columns)):
    rmse = sqrt(mean_squared_error(test.bitcoin, y_promedio[y_promedio.columns[i]]))
    print("RMSE " + y_promedio.columns[i] + ": " + str(rmse))
    rmses_y.append(rmse)
    rmses_x.append(y_promedio.columns[i])  
    
#Graficar RMSE
width = 1/1.5
plt.figure(figsize=(20,5))
plt.bar(rmses_x, rmses_y, width)
plt.xticks(rotation=20)
plt.title("RMSE",fontsize=20)
plt.show()  
```

El modelo no mejoró el RMSE. 

Método "Promedio móvil": es una mejora sobre el método de promedio simple, que toma solo el promedio de los precios en un período reciente (ventana)

```python
#Método
y_promedio_movil = test.copy()
for f in dia_futuro:
    y_promedio_movil['promedio_' + str(f) + "_movil_10"] = np.nan
    y_promedio_movil['promedio_' + str(f) + "_movil_60"] = np.nan
    for i in range(0,test_size):    
        training = df[0:len(df)-test_size+i-f+1]
        y_promedio_movil['promedio_' + str(f) + "_movil_10"][i] = training.rolling(10).mean().iloc[-1]
        y_promedio_movil['promedio_' + str(f) + "_movil_60"][i] = training.rolling(60).mean().iloc[-1]

#Graficar la serie completa
plt.figure(figsize=(20,8))
plt.plot(train.index, train['bitcoin'], label='Train')
plt.plot(test.index, test['bitcoin'], label='Test')
for i in range(1,len(y_promedio_movil.columns)):
    plt.plot(y_promedio_movil.index, y_promedio_movil[y_promedio_movil.columns[i]], label=y_promedio_movil.columns[i])
plt.legend(loc='best')
plt.title("Predicción precio Bitcoin (Promedio Móvil)",fontsize=20)
plt.show()

#Graficar solo la porción de test
plt.figure(figsize=(20,8))
plt.plot(test.index, test['bitcoin'], label='Test')
for i in range(1,len(y_promedio_movil.columns)):
    plt.plot(y_promedio_movil.index, y_promedio_movil[y_promedio_movil.columns[i]], label=y_promedio_movil.columns[i])
plt.legend(loc='best')
plt.title("Predicción precio Bitcoin (Promedio Móvil) zoom",fontsize=20)
plt.show()

#Gráficar por día futuro
plt.figure(figsize=(20,5*len(dia_futuro)/2))
col = 1
for i in range(1,len(y_promedio_movil.columns),2):
    plt.subplot(2, 2, col)
    plt.plot(test.index, test['bitcoin'], label='Test')
    plt.plot(y_promedio_movil.index, y_promedio_movil[y_promedio_movil.columns[i]], label=y_promedio_movil.columns[i])    
    plt.plot(y_promedio_movil.index, y_promedio_movil[y_promedio_movil.columns[i+1]], label=y_promedio_movil.columns[i+1])    
    plt.title("Predicción precio Bitcoin (Promedio Móvil) zoom",fontsize=20)
    plt.legend()
    plt.grid(color='grey', linestyle='-', linewidth=0.5)
    col+=1
plt.show()                     

#Cálculo de error
rmses_y = []
rmses_x = []
for i in range(1,len(y_promedio_movil.columns)):
    rmse = sqrt(mean_squared_error(test.bitcoin, y_promedio_movil[y_promedio_movil.columns[i]]))
    print("RMSE " + y_promedio_movil.columns[i] + ": " + str(rmse))
    rmses_y.append(rmse)
    rmses_x.append(y_promedio_movil.columns[i])  
    
#Graficar RMSE
width = 1/1.5
plt.figure(figsize=(20,5))
plt.bar(rmses_x, rmses_y, width)
plt.xticks(rotation=20)
plt.title("RMSE",fontsize=20)
plt.show()  
```

Se puede ver que hasta ahora el método "naive" es el que mejor ajusta para este conjunto de datos.
Un mejor acercamiento sería darle un peso a cada observación, para que las más cercanas en el tiempo influyan más en el promedio que las más lejanas.

Método "Suavizamiento Exponencial Simple": cada observación utilizada en el promedio tiene un peso asociado que decrementa exponencialmente a medida que se aleja en el tiempo (las observaciones más viejas tienen los pesos más pequeños). Modificando el factor "alfa" en la fórmula se ajusta el promedio y el error en la prediccion.

```python
#Método
y_SES = test.copy()
for f in dia_futuro:
    y_SES['SES_' + str(f) + "_smoothing_02"] = np.nan
    y_SES['SES_' + str(f) + "_smoothing_06"] = np.nan
    y_SES['SES_' + str(f) + "_smoothing_09"] = np.nan
    for i in range(0,test_size):    
        training = df[0:len(df)-test_size+i-f+1]
        ses02 = SimpleExpSmoothing(np.asarray(training)).fit(smoothing_level=0.2,optimized=False)
        ses06 = SimpleExpSmoothing(np.asarray(training)).fit(smoothing_level=0.6,optimized=False)
        ses09 = SimpleExpSmoothing(np.asarray(training)).fit(smoothing_level=0.9,optimized=False)
        y_SES['SES_' + str(f) + "_smoothing_02"][i] = ses02.forecast(1)
        y_SES['SES_' + str(f) + "_smoothing_06"][i] = ses06.forecast(1)
        y_SES['SES_' + str(f) + "_smoothing_09"][i] = ses09.forecast(1)
               
#Graficar la serie completa
plt.figure(figsize=(20,8))
plt.plot(train.index, train['bitcoin'], label='Train')
plt.plot(test.index, test['bitcoin'], label='Test')
for i in range(1,len(y_SES.columns)):
    plt.plot(y_SES.index, y_SES[y_SES.columns[i]], label=y_SES.columns[i])
plt.legend(loc='best')
plt.title("Predicción precio Bitcoin (SES)",fontsize=20)
plt.show()

#Graficar solo la porción de test
plt.figure(figsize=(20,8))
plt.plot(test.index, test['bitcoin'], label='Test')
for i in range(1,len(y_SES.columns)):
    plt.plot(y_SES.index, y_SES[y_SES.columns[i]], label=y_SES.columns[i])
plt.legend(loc='best')
plt.title("Predicción precio Bitcoin (SES) zoom",fontsize=20)
plt.show()

#Gráficar por día futuro
plt.figure(figsize=(20,5*len(dia_futuro)/2))
col = 1
for i in range(1,len(y_SES.columns),3):
    plt.subplot(2, 2, col)
    plt.plot(test.index, test['bitcoin'], label='Test')
    plt.plot(y_SES.index, y_SES[y_SES.columns[i]], label=y_SES.columns[i])   
    plt.plot(y_SES.index, y_SES[y_SES.columns[i+1]], label=y_SES.columns[i+1])
    plt.plot(y_SES.index, y_SES[y_SES.columns[i+2]], label=y_SES.columns[i+2])
    plt.title("Predicción precio Bitcoin (SES) zoom",fontsize=20)
    plt.legend()
    plt.grid(color='grey', linestyle='-', linewidth=0.5)
    col+=1
plt.show()                     

#Cálculo de error
rmses_y = []
rmses_x = []
for i in range(1,len(y_SES.columns)):
    rmse = sqrt(mean_squared_error(test.bitcoin, y_SES[y_SES.columns[i]]))
    print("RMSE " + y_SES.columns[i] + ": " + str(rmse))
    rmses_y.append(rmse)
    rmses_x.append(y_SES.columns[i])  
    
#Graficar RMSE
width = 1/1.5
plt.figure(figsize=(20,5))
plt.bar(rmses_x, rmses_y, width)
plt.xticks(rotation=20)
plt.title("RMSE",fontsize=20)
plt.show()  
```

Se puede ver que el mejor resultado lo da alfa=0.2. Sin embargo no se mejora el error del método "naive".
En general, estos métodos no funcionan bien para conjuntos de datos con mucha variabilidad.
Para conjuntos de datos que siguen una tendencia, es decir, que tienen un patrón general de movimiento de subida o de bajada (como en este caso), se puede usar el método de "Tendencia Lineal de Holt": es una extensión del método de suavizamiento exponencial simple aplicado al promedio de la serie ("nivel") y a la tendencia. La ecuación de nivel y la de tendencia se suman para general la ecuación de predicción. La suma es apropiada para conjuntos de datos con una tendencia que incrementa o decrementa linealmente, aunque también se puede usar la multiplicación de las ecuaciones de nivel y tendencia si el conjunto de datos muestra una tendencia exponencial. Se compararán los dos métodos.

```python
#Método
y_holt = test.copy()
for f in dia_futuro:
    y_holt['holt_' + str(f) + '_lineal'] = np.nan
    y_holt['holt_' + str(f) + '_exp'] = np.nan
    for i in range(0,test_size):    
        training = df[0:len(df)-test_size+i-f+1]
        holt_lineal = Holt(np.asarray(training)).fit(smoothing_level = 0.3,smoothing_slope = 0.1)
        holt_exp = Holt(np.asarray(training), exponential=True).fit(smoothing_level = 0.3,smoothing_slope = 0.1)
        y_holt['holt_' + str(f) + '_lineal'][i] = holt_lineal.forecast(1)
        y_holt['holt_' + str(f) + '_exp'][i] = holt_exp.forecast(1)
               
#Graficar la serie completa
plt.figure(figsize=(20,8))
plt.plot(train.index, train['bitcoin'], label='Train')
plt.plot(test.index, test['bitcoin'], label='Test')
for i in range(1,len(y_holt.columns)):
    plt.plot(y_holt.index, y_holt[y_holt.columns[i]], label=y_holt.columns[i])
plt.legend(loc='best')
plt.title("Predicción precio Bitcoin (Holt)",fontsize=20)
plt.show()

#Graficar solo la porción de test
plt.figure(figsize=(20,8))
plt.plot(test.index, test['bitcoin'], label='Test')
for i in range(1,len(y_holt.columns)):
    plt.plot(y_holt.index, y_holt[y_holt.columns[i]], label=y_holt.columns[i])
plt.legend(loc='best')
plt.title("Predicción precio Bitcoin (Holt) zoom",fontsize=20)
plt.show()

#Gráficar por día futuro
plt.figure(figsize=(20,5*len(dia_futuro)/2))
col = 1
for i in range(1,len(y_holt.columns),2):
    plt.subplot(2, 2, col)
    plt.plot(test.index, test['bitcoin'], label='Test')
    plt.plot(y_holt.index, y_holt[y_holt.columns[i]], label=y_holt.columns[i])   
    plt.plot(y_holt.index, y_holt[y_holt.columns[i+1]], label=y_holt.columns[i+1])
    plt.title("Predicción precio Bitcoin (Holt) zoom",fontsize=20)
    plt.legend()
    plt.grid(color='grey', linestyle='-', linewidth=0.5)
    col+=1
plt.show()                     

#Cálculo de error
rmses_y = []
rmses_x = []
for i in range(1,len(y_holt.columns)):
    rmse = sqrt(mean_squared_error(test.bitcoin, y_holt[y_holt.columns[i]]))
    print("RMSE " + y_holt.columns[i] + ": " + str(rmse))
    rmses_y.append(rmse)
    rmses_x.append(y_holt.columns[i])  
    
#Graficar RMSE
width = 1/1.5
plt.figure(figsize=(20,5))
plt.bar(rmses_x, rmses_y, width)
plt.xticks(rotation=20)
plt.title("RMSE",fontsize=20)
plt.show()  
```

Se puede ver que este método no sirve para este conjunto de datos ya que la tendencia cambia con frecuencia, no sigue un patrón.
Se puede intentar el método de Holt-Winter, que toma en cuenta la temporada en el cálculo (por ej. las visitas a un hotel en la playa van a ser influenciadas por la estación del año).
A las ecuaciones de nivel y de tendencia, se suma la ecuación de estacionalidad. Como en el método de Holt, las ecuaciones pueden sumarse o multiplicarse, prefiriendo la multiplicación cuando las variaciones estacionales cambian con el nivel de la serie (en este caso parece ser así).

```python
#Método
y_holtwinters = test.copy()
for f in dia_futuro:
    y_holtwinters['holtwinters_' + str(f) + '_lineal'] = np.nan
    y_holtwinters['holtwinters_' + str(f) + '_exp'] = np.nan
    for i in range(0,test_size):    
        training = df[0:len(df)-test_size+i-f+1]
        holt_lineal = ExponentialSmoothing(np.asarray(training), seasonal_periods=45, trend='add', seasonal='add').fit()
        holt_exp = ExponentialSmoothing(np.asarray(training), seasonal_periods=45, trend='mul', seasonal='mul').fit()
        y_holtwinters['holtwinters_' + str(f) + '_lineal'][i] = holt_lineal.forecast(1)
        y_holtwinters['holtwinters_' + str(f) + '_exp'][i] = holt_exp.forecast(1)
               
'''
#Graficar la serie completa
plt.figure(figsize=(20,8))
plt.plot(train.index, train['bitcoin'], label='Train')
plt.plot(test.index, test['bitcoin'], label='Test')
for i in range(1,len(y_holtwinters.columns)):
    plt.plot(y_holt.index, y_holtwinters[y_holtwinters.columns[i]], label=y_holtwinters.columns[i])
plt.legend(loc='best')
plt.title("Predicción precio Bitcoin (Holt)",fontsize=20)
plt.show()

#Graficar solo la porción de test
plt.figure(figsize=(20,8))
plt.plot(test.index, test['bitcoin'], label='Test')
for i in range(1,len(y_holtwinters.columns)):
    plt.plot(y_holt.index, y_holtwinters[y_holtwinters.columns[i]], label=y_holtwinters.columns[i])
plt.legend(loc='best')
plt.title("Predicción precio Bitcoin (Holt) zoom",fontsize=20)
plt.show()

#Gráficar por día futuro
plt.figure(figsize=(20,5*len(dia_futuro)/2))
col = 1
for i in range(1,len(y_holtwinters.columns),2):
    plt.subplot(2, 2, col)
    plt.plot(test.index, test['bitcoin'], label='Test')
    plt.plot(y_holtwinters.index, y_holtwinters[y_holtwinters.columns[i]], label=y_holtwinters.columns[i])   
    plt.plot(y_holtwinters.index, y_holtwinters[y_holtwinters.columns[i+1]], label=y_holtwinters.columns[i+1])
    plt.title("Predicción precio Bitcoin (Holt) zoom",fontsize=20)
    plt.legend()
    plt.grid(color='grey', linestyle='-', linewidth=0.5)
    col+=1
plt.show()                     

#Cálculo de error
rmses_y = []
rmses_x = []
for i in range(1,len(y_holtwinters.columns)):
    rmse = sqrt(mean_squared_error(test.bitcoin, y_holtwinters[y_holtwinters.columns[i]]))
    print("RMSE " + y_holtwinters.columns[i] + ": " + str(rmse))
    rmses_y.append(rmse)
    rmses_x.append(y_holtwinters.columns[i])  
    
#Graficar RMSE
width = 1/1.5
plt.figure(figsize=(20,5))
plt.bar(rmses_x, rmses_y, width)
plt.xticks(rotation=20)
plt.title("RMSE",fontsize=20)
plt.show()  
'''
```

Se eligió seasonal_periods=45 porque se puede notar que las tendencias se repiten más o menos cada 45 días. Se puede ver que este método mejora mucho el anterior. 
Aunque el RMSE sigue siendo más alto que el método naive, se puede ver que la curva de predicción del método Holt exponencial se aproxima mucho más a la curva real, y esto puede resultar mucho más significativo para predecir el movimiento del precio. 

Se probará el modelo ARIMA (Autoregressive Integrated Moving Average): intenta describir la correlación entre los datos y encontrar patrones para una predicción hacia el futuro. El método puede mejorarse utilizando estacionalidad: Seasonal ARIMA. 

```python
y_ARIMA = test.copy()
arima = sm.tsa.statespace.SARIMAX(train['bitcoin']).fit()
y_ARIMA['arima'] = arima.predict(start="2018-04-26", end="2018-05-25", dynamic=True)

plt.figure(figsize=(20,8))
plt.plot(train.index, train['bitcoin'], label='Train')
plt.plot(test.index, test['bitcoin'], label='Test')
plt.plot(y_ARIMA.index, y_ARIMA['arima'], label='Método ARIMA')
plt.legend(loc='best')
plt.title("Predicción precio Bitcoin")
plt.show()

rms = sqrt(mean_squared_error(test.bitcoin, y_ARIMA.arima))
print("RMSE (ARIMA):" + str(rms))
```

Se concluye entonces que el método ARIMA es el que menor error consigue, mientras que el método Holt-Winters multiplicativo es el que mejor ajusta la curva de predicción a la curva real.
